```python
import numpy
import matplotlib.pyplot as plt
from scipy.optimize import bisect, newton
```

```python
f1 = lambda x: x ** 5 - 1
f2 = lambda x: -numpy.sin(3 * x + 2) + x
f3 = lambda x: numpy.log(x ** 3) - 1

def closed_method(f, a, b, precision=1e-5):
    iterations = int(numpy.ceil(numpy.log2((b - a) / precision)))
    root = bisect(f, a, b, xtol=precision)
    print(f"Iterací pro přesnost 1e-5: {iterations}")
    return root

def open_method(f, x0, precision=1e-5):
    iterations = 0
    root = newton(f, x0, tol=precision)
    print(f"Iterací pro přesnost 1e-5: {iterations}")
    return root

def plot_result(f, a, b, root, title):
    print(f"Kořen funkce: {root}")
    print(f"Funkční hodnota v kořeni: {f(root)}")
    x = numpy.linspace(a, b, 100)
    plt.plot(x, f(x), "r")
    plt.plot(root, f(root), "bo")
    plt.title(title)
    plt.show()

print("\x1b[31mOhraničená metoda\x1b[0m", end="\n")

root = closed_method(f1, 0.1, 5)
plot_result(f1, 0.1, 10, root, "x**5 - 1")

root = closed_method(f2, 0.1, 1)
plot_result(f2, 0.1, 10, root, "-sin(3*x+2) + x")

root = closed_method(f3, 1, 3)
plot_result(f3, 0.1, 10, root, "log(x**3) - 1")

print("\x1b[31mNeohraničená metoda\x1b[0m", end="\n")

root = open_method(f1, 2)
plot_result(f1, 0.1, 10, root, "x**5 - 1")

root = open_method(f2, 0.1)
plot_result(f2, 0.1, 10, root, "-sin(-3*x) + x")

root = open_method(f3, 2)
plot_result(f3, 0.1, 10, root, "log(x**3) - 1")
```
